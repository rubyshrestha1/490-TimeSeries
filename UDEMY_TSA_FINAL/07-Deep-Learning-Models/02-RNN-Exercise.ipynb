## Train Test Split

**TASK: Figure out the length of the data set**

```python
# CODE HERE
len(df)
```

```python
588-12
```

**TASK: Split the data into a train/test split where the test set is the last 12 months of data.**

```python
# CODE HERE
train = df[:576]
test = df[576:]
```

```python
len(test)
```

## Scale Data

**TASK: Use a MinMaxScaler to scale the train and test sets into scaled versions.**

```python
# CODE HERE
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform (train)
scaled_test = scaler.transform (test)
```

# Time Series Generator

**TASK: Create a TimeSeriesGenerator object based off the scaled_train data. The n_input is up to you, but at a minimum it should be at least 12.**

```python
#CODE HERE
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
n_input = 12
n_features = 1
generator = TimeseriesGenerator (scaled_train, scaled_train, length = n_input, batch_size=1)
```

```python
scaled_train
```

### Create the Model

**TASK: Create a Keras Sequential Model with as many LSTAM units you want and a final Dense Layer.**

```python
len(generator)
```

```python
len(scaled_train)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(150, activation = "relu", input_shape= (n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer = "adam", loss ="mse")
```

```python
model.summary()
```

**TASK: Fit the model to the generator (it should be a lot of epochs, but do as many as you have the patience for! :)**

```python
# CODE HERE
model.fit_generator(generator, epochs = 20)
```

**TASK: Plot the history of the loss that occured during training.**

```python
# CODE HERE
```

```python
model.history.history.keys()
```

```python
plt.plot(range(len(model.history.history['loss'])), model.history.history['loss'])
```

**TASK: Generate predictions into the same time stamps as the test set**

```python
# CODE HERE
```

```python
eval_batch = scaled_train[-n_input:]
current_batch = eval_batch.reshape((1, n_input, n_features))

test_predictions=[]
for i in range (len(scaled_test)):
    prediction = model.predict (current_batch)[0]
    test_predictions.append(prediction)
    current_batch = np.append(current_batch[:, 1:, :], [[prediction]], axis=1)
test_predictions
```

```python
scaled_test
```

```python
plt.plot(range(len(scaled_test)), scaled_test)
plt.plot(range(len(scaled_test)), test_predictions)
```
